```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# import data to a dataframe

url = 'https://github.com/prasertcbs/basic-dataset/raw/master/boston_house_price.csv'
df = pd.read_csv(url)
df.head(3)
```

```python
df.info()
```

```python
# plot data distribution

df.hist(figsize = (15, 10));
```

### sklearn preprocessing

```python
# import library

from sklearn import preprocessing
```

```python
# cereate StandardScaler instance

scaler = preprocessing.StandardScaler() 
scaler
```

```python
# compute the mean and std

scaler.fit(df[['RM']])
```

```python
# calculate z-score

z = scaler.transform(df[['RM']])
z
```

```python
z[:5]
```

```python
print(f'mean: {z.mean()}')
print(f'std: {z.std()}')
```

```python
df.head(3)
```

```python
# create a new column to store r_RM

df['r_RM'] = z[:, 0]
df.head(3)
```

```python
cols = ['RM', 'MEDV']
scaler = preprocessing.StandardScaler()
scaler
```

```python
scaler.fit(df[cols])
```

```python
z = scaler.transform(df[cols])
z
```

```python
z.mean(axis = 0)
```

```python
z.std(axis = 0)
```

```python
# create new 2 columns to store data

ds = pd.concat([df, pd.DataFrame(z, columns = ['z_room', 'z_medv'])], axis = 'columns')
ds.head(3)
```
